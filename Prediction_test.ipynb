```python

# ▶ Import libraries
import pickle
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

nltk.download('stopwords', quiet=True)

print('✅ Libraries loaded!')
```

```python

# ▶ Load the saved model and vectorizer
with open('model.pkl', 'rb') as f:
    model = pickle.load(f)

with open('tfidf_vectorizer.pkl', 'rb') as f:
    tfidf = pickle.load(f)

print('✅ Model and vectorizer loaded successfully!')
```

```python

# ▶ Define the same cleaning function (must match Step 2 exactly)
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    words = text.split()
    words = [stemmer.stem(w) for w in words if w not in stop_words and len(w) > 2]
    return ' '.join(words)

def predict_news(news_text):
    """Takes raw news text and returns prediction with confidence."""
    cleaned = clean_text(news_text)
    vectorized = tfidf.transform([cleaned])
    prediction = model.predict(vectorized)[0]
    probability = model.predict_proba(vectorized)[0]

    label = '✅ REAL NEWS' if prediction == 1 else '🚨 FAKE NEWS'
    confidence = probability[prediction] * 100

    print(f'Prediction : {label}')
    print(f'Confidence : {confidence:.1f}%')
    print(f'Fake probability : {probability[0]*100:.1f}%')
    print(f'Real probability : {probability[1]*100:.1f}%')
    return prediction, confidence

print('✅ Prediction function ready!')
```

```python

# ▶ TEST 1: Example of likely FAKE news
fake_example = """
BREAKING: Scientists discover that drinking bleach cures all diseases.
The government is hiding this miracle cure from the public.
Share this before they delete it! Deep state cover-up exposed!
"""
print('--- TEST 1: Likely Fake News ---')
predict_news(fake_example)
```

```python

# ▶ TEST 2: Example of likely REAL news
real_example = """
The Federal Reserve raised interest rates by 0.25 percentage points on Wednesday,
its tenth consecutive increase as officials continue efforts to bring inflation
back to the central bank's 2% target. The decision was unanimous among policymakers.
Chair Jerome Powell said the committee would assess incoming data carefully.
"""
print('--- TEST 2: Likely Real News ---')
predict_news(real_example)
```
